```julia
function iterator(f::Function, n::Int, x0)
    
    solution = Float64[x0]
    steps = Int[0]
    x_old = x0
    for i in 1:n
        
        x_new = f(solution[i])
        push!(solution, x_new)
        push!(steps, i)
        
    end
    
    return steps, solution
    
end
```

```julia
f(x) = 3(x -x^2)
```

```julia
N, tray = iterator(f, 100, 0.9)
plot(N, tray)
```
