```julia
function logistic_equation(x0, r)
    
    x = r*x0*(1-x0)
    return x
    
end
```

```julia
function logistic_iterator(x0, r, n, yn::Bool)
    
    solutions = []
    times = collect(0:n)
    
    push!(solutions, x0)
    
    for i in 1:n
        
        x_new = logistic_equation(x0, r)
        push!(solutions, x_new)
        x0 = x_new
        
    end
    
    plot(times, solutions)
    
    if yn == true
    
        plot(times, solutions, ".")
        
    end
    
    println(solutions[n+1])
    
end 
```

```julia
logistic_equation(0.5, 1.5)
```

```julia
logistic_iterator(0.5, 1.5, 20)
```

```julia
logistic_iterator(0.4, 3.5, 50)
```
