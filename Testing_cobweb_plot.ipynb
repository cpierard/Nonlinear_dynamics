```julia
function cobweb_plot(f::Function, x0, rangex, n)

    xx = [x0]
    fx = [0.0]
    for it = 1:n

          x1 = f(x0)
          push!(xx, x0)
          push!(fx, x1)
          x0 = x1
          push!(xx,x1)
          push!(fx,x1)

      end

    plot(rangex, f, xaxis=(L"x", (rangex[1], rangex[end])), yaxis=L"F(x)=$f")
    plot!(rangex, identity)
    plot!(xx, fx, marker=(:dot, 3, 0.4))

end
```

```julia
latexstring("an equation: \$1 + \\alpha^2\$")
```

```julia
f(x) = 3.8(x - x^2)
```

```julia
 cobweb_plot(f, 0.1, 0:1/32:1, 20)
```

```julia
@manipulate for k=0:0.1:1
    
    cobweb_plot(f, k, 0:1/32:1, 20)
    
end
```
